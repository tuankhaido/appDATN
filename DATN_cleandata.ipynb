```python
import pandas as pd

# Tải file Excel
file_path = 'KhungChuongTrinh-CNTT_K62.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Khởi tạo danh sách để lưu trữ thông tin trích xuất
ten_hoc_phan = []
ma_hoc_phan = []
so_tin_chi = []
hoc_ky = []

# Biến để theo dõi học kỳ hiện tại
current_semester = None

# Duyệt qua từng hàng trong DataFrame
for index, row in df.iterrows():
    # Kiểm tra nếu hàng chỉ ra học kỳ mới
    if 'HỌC KỲ' in str(row.iloc[1]):
        current_semester = row.iloc[1].split()[-1]  # Lấy giá trị học kỳ từ chuỗi "HỌC KỲ X"
    elif pd.notna(row.iloc[1]) and pd.notna(row.iloc[2]) and pd.notna(row.iloc[3]) and isinstance(row.iloc[3], (int, float)):
        # Trích xuất tên môn học và bỏ các ký tự không cần thiết
        ten_hoc_phan.append(row.iloc[1])
        ma_hoc_phan.append(row.iloc[2])
        so_tin_chi.append(row.iloc[3])
        hoc_ky.append(current_semester)

# Tạo DataFrame mới với thông tin đã trích xuất
cleaned_data = pd.DataFrame({
    'Tên Học Phần': ten_hoc_phan,
    'Mã Học Phần': ma_hoc_phan,
    'Số Tín Chỉ': so_tin_chi,
    'Học Kỳ': hoc_ky
})
#Viết hoa các chữ cái đầu trong Tên Học Phần
cleaned_data['Tên Học Phần']= cleaned_data['Tên Học Phần'].str.title()
# Hàm để làm sạch tên học phần
def clean_subject_name(subject_name):
    return str(subject_name).replace("A-", "").replace("B-", "").replace("C-", "").replace("D-", "").replace("E-", "").strip()
# Áp dụng hàm làm sạch cho cột 'Tên Học Phần'
cleaned_data['Tên Học Phần'] = cleaned_data['Tên Học Phần'].apply(clean_subject_name)
cleaned_data.loc[cleaned_data['Mã Học Phần'] == 'PS0.001.3', 'Tên Học Phần'] = 'Triết Học Mác- Lênin'
cleaned_data.loc[cleaned_data['Mã Học Phần'] == 'IT1.222.3', 'Tên Học Phần'] = 'An Toàn Bảo Mật Thông Tin'
cleaned_data.loc[cleaned_data['Mã Học Phần'] == 'CPM10.3', 'Tên Học Phần'] = 'Chuyên Đề Về Công Nghệ Phần Mềm'
cleaned_data.loc[cleaned_data['Mã Học Phần'] == 'PS0.002.2', 'Tên Học Phần'] = 'Kinh Tế Chính Trị Mác- Lênin'
cleaned_data.loc[cleaned_data['Mã Học Phần'] == 'ANHCNTT.3', 'Tên Học Phần'] = 'Tiếng Anh Chuyên Ngành'
cleaned_data = cleaned_data[~cleaned_data['Tên Học Phần'].str.contains('Giáo Dục Thể Chất', na=False)]
cleaned_data = cleaned_data[~cleaned_data['Tên Học Phần'].str.contains('Giáo Dục Qp-An', na=False)]
# Reset lại chỉ số
cleaned_data.reset_index(drop=True, inplace=True)
# Lưu dữ liệu đã làm sạch vào file Excel mới
cleaned_data.to_excel('Cleaned_KhungChuongTrinh_K62.xlsx', index=False)
pd.set_option('display.max_rows', None)
# Hiển thị kết quả
print(cleaned_data)
```

```python
import pandas as pd

# Đọc hai file Excel
file_k60 = 'Cleaned_KhungChuongTrinh_K60.xlsx'
file_k62 = 'Cleaned_KhungChuongTrinh_K62.xlsx'

df_k60 = pd.read_excel(file_k60, engine='openpyxl')
df_k62 = pd.read_excel(file_k62, engine='openpyxl')

# Tạo một DataFrame chứa các tên môn học từ K62
existing_courses = df_k62['Tên Học Phần'].tolist()

# Lọc ra các môn học từ K60 mà không có trong K62
df_k60_new = df_k60[~df_k60['Tên Học Phần'].isin(existing_courses)]

# Kết hợp DataFrame K62 với các môn học mới từ K60
df_combined = pd.concat([df_k62, df_k60_new], ignore_index=True)

# Đảm bảo giữ mã môn học của K62 nếu có môn cùng tên
df_combined = df_combined.groupby('Tên Học Phần').first().reset_index()

# Chuyển đổi cột "Học Kỳ" sang kiểu số để có thể sắp xếp đúng
df_combined['Học Kỳ'] = pd.to_numeric(df_combined['Học Kỳ'], errors='coerce')

# Sắp xếp theo thứ tự học kỳ
df_combined = df_combined.sort_values(by='Học Kỳ').reset_index(drop=True)

# Lưu DataFrame kết hợp vào một file Excel mới
output_file = 'Integrated_KhungChuongTrinh.xlsx'
df_combined.to_excel(output_file, index=False, engine='openpyxl')

pd.set_option('display.max_rows', None)
# Hiển thị kết quả
print(df_combined)
```

```python
import pandas as pd

# --- Hàm chuẩn hóa dữ liệu ---
def normalize_value(value):
    if isinstance(value, str):
        value = value.replace(',', '.')
        if value.replace('.', '', 1).isdigit():
            value = float(value)
    if isinstance(value, (int, float)):
        value = round(float(value), 2)
    return value

# --- Đọc tất cả các sheet từ file Excel ---
file_path = 'Chua_TN.xlsx'
xls = pd.ExcelFile(file_path, engine='openpyxl')

all_merged_data = []

for sheet_name in xls.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name, engine='openpyxl')
    df = df.map(normalize_value)
    df = df.drop([0, 1, 2, 3, 4, 6, 7, 8]).reset_index(drop=True)

    # --- Gộp cột L1 và L2 ---
    def merge_l1_l2(row):
        merged_values = []
        num_cols_to_exclude = 10
        for col in range(2, len(df.columns) - num_cols_to_exclude, 2):
            l1 = row.iloc[col]
            l2 = row.iloc[col + 1] if col + 1 < len(df.columns) - num_cols_to_exclude else None
            merged_value = ', '.join(str(x) for x in [l1, l2] if pd.notna(x)) or None
            merged_values.append(merged_value)
        return merged_values

    merged_data = []
    for _, row in df.iterrows():
        merged_data.append([row.iloc[1]] + merge_l1_l2(row))
    
    all_merged_data.extend(merged_data)

# Ensure the header matches the number of columns in rows
header = all_merged_data[0]
rows = all_merged_data[1:]

# Adjust the header length to match the rows if necessary
if len(header) > len(rows[0]):
    header = header[:len(rows[0])]
elif len(header) < len(rows[0]):
    rows = [row[:len(header)] for row in rows]

# Create DataFrame with adjusted header and rows
merged_df = pd.DataFrame(rows, columns=header)
cleaned_df = pd.DataFrame(rows, columns=header)
cleaned_df = merged_df.dropna(how='all', axis=1)
cleaned_df.rename(columns={'Mã\nsinh viên': 'Mã sinh viên'}, inplace=True)

# Ensure 'Mã sinh viên' is numeric
cleaned_df['Mã sinh viên'] = pd.to_numeric(cleaned_df['Mã sinh viên'].astype(str), errors='coerce').fillna(0).astype(int).astype(str)

# Xóa các hàng mà cột 'Mã sinh viên' = 0
cleaned_df = cleaned_df[cleaned_df['Mã sinh viên'] != '0']
cleaned_df.columns = cleaned_df.columns.str.title()

print(cleaned_df)
cleaned_df.to_excel('Cleaned_ChuaTotNghiep.xlsx', index=False)

```

```python
from typing import final
import pandas as pd
import re
import os

# Khởi tạo danh sách để lưu DataFrame đã xử lý cho từng file
all_combined_data = []

# Lấy danh sách tất cả các file trong thư mục 'Bang diem'
directory_path = './Bang diem'
file_list = [f for f in os.listdir(directory_path) if f.endswith('.xlsx')]

for file_name in file_list:
    file_path = os.path.join(directory_path, file_name)
    xls = pd.ExcelFile(file_path)  # Đọc file Excel

    # Khởi tạo danh sách để lưu DataFrame đã xử lý cho từng sheet
    all_data = []
    student_ratings = {}  # Dictionary để lưu xếp loại của mỗi sinh viên

    # Lặp qua tất cả các sheet trong file
    for sheet_name in xls.sheet_names:
        df = pd.read_excel(file_path, sheet_name=sheet_name)  # Đọc từng sheet
        df = df.drop([0, 1, 2, 3, 4])  # Xóa các hàng không cần thiết
        df.reset_index(drop=True, inplace=True)  # Đặt lại chỉ số

        # Lấy giá trị "Xếp loại" từ dòng 62, cột 0
        rating_text = str(df.iloc[55, 1])
        rating_match = re.search(r'Xếp loại: (\w+)', rating_text)
        rating = rating_match.group(1) if rating_match else 'Không xác định'
        student_ratings[sheet_name] = rating

        # Tạo danh sách để lưu trữ dữ liệu sạch
        data = []

        # Lấy kết quả học tập
        courses = df.iloc[10:54, [2, 7, 8]]  # Lấy cột Học phần, Điểm L1 và L2
        courses.columns = ['Học phần', 'Điểm L1', 'Điểm L2']  # Đặt tên cột

        # Gộp các giá trị điểm L1 và L2 vào một ô
        for index, row in courses.iterrows():
            merged_row = [sheet_name, row['Học phần']]

            # Gộp điểm L1 và L2
            l1_value = row['Điểm L1']
            l2_value = row['Điểm L2']

            # Áp dụng logic gộp
            if pd.notnull(l1_value) and pd.notnull(l2_value):
                merged_value = f"{l1_value}, {l2_value}"  # Gộp với dấu ", "
            elif pd.notnull(l1_value):
                merged_value = l1_value  # Chỉ L1
            elif pd.notnull(l2_value):
                merged_value = l2_value  # Chỉ L2
            else:
                merged_value = 'Null'  # Cả hai đều null

            merged_row.append(merged_value)  # Thêm giá trị gộp vào danh sách
            data.append(merged_row)

        # Tạo DataFrame mới từ các hàng đã gộp
        cleaned_df = pd.DataFrame(data, columns=["Mã Sinh Viên", "Học phần", "Điểm"])

        # Kết hợp dữ liệu từ tất cả các sheet
        all_data.append(cleaned_df)

    # Kết hợp tất cả các DataFrame đã xử lý thành một DataFrame
    combined_df = pd.concat(all_data, ignore_index=True)

    # Đặt thứ tự cho cột Học phần
    combined_df['Học phần'] = pd.Categorical(combined_df['Học phần'], categories=combined_df['Học phần'].unique(), ordered=True)

    # Thêm môn học từ khung chương trình vào DataFrame
    courses_file_path = 'Integrated_KhungChuongTrinh.xlsx'
    courses_df = pd.read_excel(courses_file_path, engine='openpyxl')
    courses_df = courses_df[['Tên Học Phần']]  # Giả sử cột chứa tên môn học là "Tên Học Phần"
    courses_df.rename(columns={'Tên Học Phần': 'Học phần'}, inplace=True)

    # Chuyển đổi tên học phần trong courses_df thành chữ hoa chữ đầu tiên
    courses_df['Học phần'] = courses_df['Học phần'].str.title()

    # Tạo danh sách để lưu trữ dữ liệu cuối cùng
    final_data = []

    # Lưu thứ tự môn học vào danh sách
    course_order = courses_df['Học phần'].tolist()

    for index, row in combined_df.iterrows():
        sheet_name = row['Mã Sinh Viên']

        # Chuyển đổi tên học phần trong row thành chữ hoa chữ đầu tiên
        current_course = row['Học phần'].title()

        for course in course_order:
            # Lấy điểm cho môn học tương ứng hoặc None nếu không có
            point = row['Điểm'] if current_course == course else None
            final_data.append({
                'Mã Sinh Viên': sheet_name,
                'Học phần': course,
                'Điểm': point
            })

    # Tạo DataFrame cuối cùng
    final_df = pd.DataFrame(final_data)

    # Nhóm dữ liệu để loại bỏ các giá trị trùng lặp
    final_df = final_df.groupby(['Mã Sinh Viên', 'Học phần'], as_index=False).agg({'Điểm': 'first'})

    # Đặt thứ tự cột cho DataFrame cuối cùng
    final_df['Học phần'] = pd.Categorical(final_df['Học phần'], categories=course_order, ordered=True)

    # Xoay dọc các hàng thành cột
    pivot_df = final_df.pivot(index=['Mã Sinh Viên'], columns='Học phần', values='Điểm')

    # Đặt tên cột lại
    pivot_df.columns.name = None  # Xóa tên cột để đẹp hơn
    pivot_df.reset_index(inplace=True)  # Đặt lại chỉ số

    # Thêm cột "Xếp loại" vào DataFrame cuối cùng
    pivot_df['Xếp loại'] = pivot_df['Mã Sinh Viên'].map(student_ratings)

    # Kết hợp dữ liệu từ tất cả các file
    all_combined_data.append(pivot_df)

# Kết hợp tất cả các DataFrame đã xử lý từ tất cả các file thành một DataFrame
final_combined_df = pd.concat(all_combined_data, ignore_index=True)
final_combined_df['Xếp loại'] = final_combined_df['Xếp loại'].replace('Xuất', 'Xuất Sắc')
final_combined_df['Xếp loại'] = final_combined_df['Xếp loại'].replace('Trung', 'Trung Bình')

# Lưu dữ liệu đã xử lý vào file Excel mới
output_file_path = 'Cleaned_DaTotNghiep_All.xlsx'
final_combined_df.to_excel(output_file_path, index=False)

print("Đã hoàn thành việc tạo file mới:", output_file_path)
```

```python
import pandas as pd

# Đọc file Cleaned_DaTotNghiep.xlsx
try:
    graduated_df = pd.read_excel('Cleaned_DaTotNghiep_Final.xlsx', engine='openpyxl')
except FileNotFoundError:
    print("Không tìm thấy file 'Cleaned_DaTotNghiep.xlsx'.")
    exit()

# Đọc file Cleaned_ChuaTotNghiep.xlsx
try:
    not_graduated_df = pd.read_excel('Cleaned_ChuaTotNghiep_All.xlsx', engine='openpyxl')
except FileNotFoundError:
    print("Không tìm thấy file 'Cleaned_ChuaTotNghiep.xlsx'.")
    exit()

# Gộp hai DataFrame lại với nhau
combined_df = pd.concat([graduated_df, not_graduated_df], ignore_index=True)

# Đổi cột 'Mã sinh viên' thành dạng text
combined_df['Mã Sinh Viên'] = combined_df['Mã Sinh Viên'].astype(str)

# Hàm quy đổi điểm
def convert_grade(score):
    if pd.isna(score):  # Kiểm tra nếu ô trống
        return None
    if isinstance(score, str):
        try:
            score = float(score.split(',')[-1])  # Lấy giá trị cuối cùng nếu có nhiều giá trị
        except ValueError:
            return None  # Bỏ qua các giá trị không thể chuyển đổi thành số
    if isinstance(score, (int, float)):
        if score >= 8.5:
            return 'A'
        elif score >= 7.0:
            return 'B'
        elif score >= 5.5:
            return 'C'
        elif score >= 4.0:
            return 'D'
        else:
            return 'F'
    return None  # Trả về None nếu không phải số

# Áp dụng hàm quy đổi cho các cột điểm
for column in combined_df.columns[1:-1]:  # Bỏ qua cột 'Mã sinh viên', 'Họ và tên' và 'Xếp loại'
    combined_df[column] = combined_df[column].apply(convert_grade)
# TÍCH HỢP MÔN CHO NĂM NHẤT
# Merge 'Nhập Môn Ngành Cntt' and 'Kỹ Năng Mềm' columns
combined_df['Nhập Môn Ngành Cntt'] = combined_df['Nhập Môn Ngành Cntt'].fillna('').astype(str) + ', ' + combined_df['Kỹ Năng Mềm'].fillna('').astype(str)
combined_df['Nhập Môn Ngành Cntt'] = combined_df['Nhập Môn Ngành Cntt'].str.strip(', ')
combined_df.drop(columns=['Kỹ Năng Mềm'], inplace=True)
# Merge 'Kỹ Thuật Lập Trình' and 'Lập Trình Nâng Cao' columns
combined_df['Kỹ Thuật Lập Trình'] = combined_df['Kỹ Thuật Lập Trình'].fillna('').astype(str) + ', ' + combined_df['Lập Trình Nâng Cao'].fillna('').astype(str)
combined_df['Kỹ Thuật Lập Trình'] = combined_df['Kỹ Thuật Lập Trình'].str.strip(', ')
combined_df.drop(columns=['Lập Trình Nâng Cao'], inplace=True)
# Merge 'Bảo Trì Hệ Thống' and 'Giải Tích Số' columns
combined_df['Bảo Trì Hệ Thống'] = combined_df['Bảo Trì Hệ Thống'].fillna('').astype(str) + ', ' + combined_df['Giải Tích Số'].fillna('').astype(str)
combined_df['Bảo Trì Hệ Thống'] = combined_df['Bảo Trì Hệ Thống'].str.strip(', ')
combined_df.drop(columns=['Giải Tích Số'], inplace=True)
# TÍCH HỢP MÔN CHO NĂM 2
# Merge 'Bảo Trì Hệ Thống' and 'Giải Tích Số' columns
# Merge 'Nguyên Lý Hệ Điều Hành' and 'Hệ Điều Hành' columns
combined_df['Nguyên Lý Hệ Điều Hành'] = combined_df['Nguyên Lý Hệ Điều Hành'].fillna('').astype(str) + ', ' + combined_df['Hệ Điều Hành'].fillna('').astype(str)
combined_df['Nguyên Lý Hệ Điều Hành'] = combined_df['Nguyên Lý Hệ Điều Hành'].str.strip(', ')
combined_df.drop(columns=['Hệ Điều Hành'], inplace=True)
# Merge 'Tiếng Anh B1', 'Tiếng Nga B1', and 'Tiếng Pháp B1' columns
combined_df['Tiếng Anh B1'] = combined_df['Tiếng Anh B1'].fillna('').astype(str) + ', ' + combined_df['Tiếng Nga B1'].fillna('').astype(str) + ', ' + combined_df['Tiếng Pháp B1'].fillna('').astype(str)
combined_df['Tiếng Anh B1'] = combined_df['Tiếng Anh B1'].str.strip(', ')
combined_df.drop(columns=['Tiếng Nga B1', 'Tiếng Pháp B1'], inplace=True)
# TÍCH HỢP MÔN CHO NĂM 3
# Merge 'Tiếng Anh Chuyên Ngành', 'Tiếng Nga Chuyên Ngành', and 'Tiếng Pháp Chuyên Ngành' columns
combined_df['Tiếng Anh Chuyên Ngành'] = combined_df['Tiếng Anh Chuyên Ngành'].fillna('').astype(str) + ', ' + combined_df['Tiếng Nga Chuyên Ngành'].fillna('').astype(str) + ', ' + combined_df['Tiếng Pháp Chuyên Ngành'].fillna('').astype(str)
combined_df['Tiếng Anh Chuyên Ngành'] = combined_df['Tiếng Anh Chuyên Ngành'].str.strip(', ')
combined_df.drop(columns=['Tiếng Nga Chuyên Ngành', 'Tiếng Pháp Chuyên Ngành'], inplace=True)
# Merge 'Thuật Toán Và Ứng Dụng' and 'Chương Trình Dịch' columns
combined_df['Thuật Toán Và Ứng Dụng'] = combined_df['Thuật Toán Và Ứng Dụng'].fillna('').astype(str) + ', ' + combined_df['Chương Trình Dịch'].fillna('').astype(str)
combined_df['Thuật Toán Và Ứng Dụng'] = combined_df['Thuật Toán Và Ứng Dụng'].str.strip(', ')
combined_df.drop(columns=['Chương Trình Dịch'], inplace=True)
# Merge 'Lập Trình Thiết Bị Di Động' and 'Thực Hành Mạng' columns
combined_df['Lập Trình Thiết Bị Di Động'] = combined_df['Lập Trình Thiết Bị Di Động'].fillna('').astype(str) + ', ' + combined_df['Thực Hành Mạng'].fillna('').astype(str)
combined_df['Lập Trình Thiết Bị Di Động'] = combined_df['Lập Trình Thiết Bị Di Động'].str.strip(', ')
combined_df.drop(columns=['Thực Hành Mạng'], inplace=True)
# Merge 'Phân Tích Thiết Kế Yêu Cầu' and 'Phân Tích Thiết Kế Hệ Thống' columns
combined_df['Phân Tích Thiết Kế Yêu Cầu'] = combined_df['Phân Tích Thiết Kế Yêu Cầu'].fillna('').astype(str) + ', ' + combined_df['Phân Tích Thiết Kế Hệ Thống'].fillna('').astype(str)
combined_df['Phân Tích Thiết Kế Yêu Cầu'] = combined_df['Phân Tích Thiết Kế Yêu Cầu'].str.strip(', ')
combined_df.drop(columns=['Phân Tích Thiết Kế Hệ Thống'], inplace=True)
# Merge 'Phân Tích Thiết Kế Hướng Đối Tượng' and 'Xử Lý Ảnh' columns
combined_df['Phân Tích Thiết Kế Hướng Đối Tượng'] = combined_df['Phân Tích Thiết Kế Hướng Đối Tượng'].fillna('').astype(str) + ', ' + combined_df['Xử Lý Ảnh'].fillna('').astype(str)
combined_df['Phân Tích Thiết Kế Hướng Đối Tượng'] = combined_df['Phân Tích Thiết Kế Hướng Đối Tượng'].str.strip(', ')
combined_df.drop(columns=['Xử Lý Ảnh'], inplace=True)
# Merge 'Project 1' and 'Thực Tập Chuyên Môn' columns
combined_df['Project 1'] = combined_df['Project 1'].fillna('').astype(str) + ', ' + combined_df['Thực Tập Chuyên Môn'].fillna('').astype(str)
combined_df['Project 1'] = combined_df['Project 1'].str.strip(', ')
combined_df.drop(columns=['Thực Tập Chuyên Môn'], inplace=True)
# Merge 'Thiết Kế Cơ Sở Dữ Liệu', 'Hệ Quản Trị CSDL SQLServer', and 'Hệ Quản Trị CSDL Oracle' columns
combined_df.insert(
    combined_df.columns.get_loc('Hệ Quản Trị Csdl Sqlserver'),
    'Hệ Quản Trị CSDL SQLServer/Oracle',
    combined_df['Thiết Kế Cơ Sở Dữ Liệu'].fillna('').astype(str) + ', ' +
    combined_df['Hệ Quản Trị Csdl Sqlserver'].fillna('').astype(str) + ', ' +
    combined_df['Hệ Quản Trị Csdl Oracle'].fillna('').astype(str)
)
combined_df['Hệ Quản Trị CSDL SQLServer/Oracle'] = combined_df['Hệ Quản Trị CSDL SQLServer/Oracle'].str.strip(', ')
combined_df.drop(columns=['Thiết Kế Cơ Sở Dữ Liệu', 'Hệ Quản Trị Csdl Sqlserver', 'Hệ Quản Trị Csdl Oracle'], inplace=True)
# Merge 'Lập Trình Sử Dụng Api' and 'Công Nghệ Phần Mềm' columns
combined_df.insert(
    combined_df.columns.get_loc('Lập Trình Sử Dụng Api'),
    'Lập Trình Sử Dụng Api/Công Nghệ Phần Mềm',
    combined_df['Lập Trình Sử Dụng Api'].fillna('').astype(str) + ', ' +
    combined_df['Công Nghệ Phần Mềm'].fillna('').astype(str)
)
combined_df['Lập Trình Sử Dụng Api/Công Nghệ Phần Mềm'] = combined_df['Lập Trình Sử Dụng Api/Công Nghệ Phần Mềm'].str.strip(', ')
combined_df.drop(columns=['Lập Trình Sử Dụng Api', 'Công Nghệ Phần Mềm'], inplace=True)
# Xóa các hàng có nhiều hơn 40 cột trống
combined_df = combined_df.dropna(thresh=len(combined_df.columns) - 40)
# Xóa toàn bộ các cột không có giá trị nào
combined_df.dropna(axis=1, how='all', inplace=True)
# Xóa các cột có nhiều hơn 160 dòng trống
combined_df = combined_df.dropna(axis=1, thresh=len(combined_df) - 160)
# Xóa các cột không cần thiết
columns_to_drop = ['Khai Phá Dữ Liệu', 'Chuyên Đề Về Công Nghệ Phần Mềm', 'An Ninh Mạng']
combined_df.drop(columns=columns_to_drop, inplace=True)
# Sắp xếp theo tên sinh viên
# combined_df.sort_values(by='Mã Sinh Viên', inplace=True)
# Fill missing values in specific columns with 0
columns_to_fill = ['Tiếng Anh Chuyên Ngành', 'Tiếng Anh B1', 'Thực Tập Tốt Nghiệp', 'Đồ Án Tốt Nghiệp']
combined_df[columns_to_fill] = combined_df[columns_to_fill].fillna(0)
# Lưu kết quả vào file mới
combined_df.to_excel('Combined_Results.xlsx', index=False)

print("Đã hoàn thành việc gộp hai file và quy đổi điểm.")
```

```python
import pandas as pd

# Đọc file Cleaned_DaTotNghiep.xlsx
try:
    graduated_df = pd.read_excel('Cleaned_DaTotNghiep_Final.xlsx', engine='openpyxl')
except FileNotFoundError:
    print("Không tìm thấy file 'Cleaned_DaTotNghiep.xlsx'.")
    exit()

# Đọc file Cleaned_ChuaTotNghiep.xlsx
try:
    not_graduated_df = pd.read_excel('Cleaned_ChuaTotNghiep_All.xlsx', engine='openpyxl')
except FileNotFoundError:
    print("Không tìm thấy file 'Cleaned_ChuaTotNghiep.xlsx'.")
    exit()

# Gộp hai DataFrame lại với nhau
combined_df = pd.concat([graduated_df, not_graduated_df], ignore_index=True)

# Đổi cột 'Mã sinh viên' thành dạng text
combined_df['Mã Sinh Viên'] = combined_df['Mã Sinh Viên'].astype(str)

# Hàm quy đổi điểm
def get_higher_score(score):
    if isinstance(score, str):
        scores = [float(s) for s in score.split(',') if s.strip().replace('.', '', 1).isdigit()]
        return max(scores) if scores else None
    elif isinstance(score, (int, float)):
        return score
    else:
        return None

# Áp dụng hàm lấy điểm cao hơn cho các cột điểm
for column in combined_df.columns[1:-1]:  # Bỏ qua cột 'Mã sinh viên' và 'Xếp loại'
    combined_df[column] = combined_df[column].apply(get_higher_score)
# TÍCH HỢP MÔN CHO NĂM NHẤT
# Merge 'Nhập Môn Ngành Cntt' and 'Kỹ Năng Mềm' columns
combined_df['Nhập Môn Ngành Cntt'] = combined_df['Nhập Môn Ngành Cntt'].fillna('').astype(str) + ', ' + combined_df['Kỹ Năng Mềm'].fillna('').astype(str)
combined_df['Nhập Môn Ngành Cntt'] = combined_df['Nhập Môn Ngành Cntt'].str.strip(', ')
combined_df.drop(columns=['Kỹ Năng Mềm'], inplace=True)
# Merge 'Kỹ Thuật Lập Trình' and 'Lập Trình Nâng Cao' columns
combined_df['Kỹ Thuật Lập Trình'] = combined_df['Kỹ Thuật Lập Trình'].fillna('').astype(str) + ', ' + combined_df['Lập Trình Nâng Cao'].fillna('').astype(str)
combined_df['Kỹ Thuật Lập Trình'] = combined_df['Kỹ Thuật Lập Trình'].str.strip(', ')
combined_df.drop(columns=['Lập Trình Nâng Cao'], inplace=True)
# Merge 'Bảo Trì Hệ Thống' and 'Giải Tích Số' columns
combined_df['Bảo Trì Hệ Thống'] = combined_df['Bảo Trì Hệ Thống'].fillna('').astype(str) + ', ' + combined_df['Giải Tích Số'].fillna('').astype(str)
combined_df['Bảo Trì Hệ Thống'] = combined_df['Bảo Trì Hệ Thống'].str.strip(', ')
combined_df.drop(columns=['Giải Tích Số'], inplace=True)
# TÍCH HỢP MÔN CHO NĂM 2
# Merge 'Bảo Trì Hệ Thống' and 'Giải Tích Số' columns
# Merge 'Nguyên Lý Hệ Điều Hành' and 'Hệ Điều Hành' columns
combined_df['Nguyên Lý Hệ Điều Hành'] = combined_df['Nguyên Lý Hệ Điều Hành'].fillna('').astype(str) + ', ' + combined_df['Hệ Điều Hành'].fillna('').astype(str)
combined_df['Nguyên Lý Hệ Điều Hành'] = combined_df['Nguyên Lý Hệ Điều Hành'].str.strip(', ')
combined_df.drop(columns=['Hệ Điều Hành'], inplace=True)
# Merge 'Tiếng Anh B1', 'Tiếng Nga B1', and 'Tiếng Pháp B1' columns
combined_df['Tiếng Anh B1'] = combined_df['Tiếng Anh B1'].fillna('').astype(str) + ', ' + combined_df['Tiếng Nga B1'].fillna('').astype(str) + ', ' + combined_df['Tiếng Pháp B1'].fillna('').astype(str)
combined_df['Tiếng Anh B1'] = combined_df['Tiếng Anh B1'].str.strip(', ')
combined_df.drop(columns=['Tiếng Nga B1', 'Tiếng Pháp B1'], inplace=True)
# TÍCH HỢP MÔN CHO NĂM 3
# Merge 'Tiếng Anh Chuyên Ngành', 'Tiếng Nga Chuyên Ngành', and 'Tiếng Pháp Chuyên Ngành' columns
combined_df['Tiếng Anh Chuyên Ngành'] = combined_df['Tiếng Anh Chuyên Ngành'].fillna('').astype(str) + ', ' + combined_df['Tiếng Nga Chuyên Ngành'].fillna('').astype(str) + ', ' + combined_df['Tiếng Pháp Chuyên Ngành'].fillna('').astype(str)
combined_df['Tiếng Anh Chuyên Ngành'] = combined_df['Tiếng Anh Chuyên Ngành'].str.strip(', ')
combined_df.drop(columns=['Tiếng Nga Chuyên Ngành', 'Tiếng Pháp Chuyên Ngành'], inplace=True)
# Merge 'Thuật Toán Và Ứng Dụng' and 'Chương Trình Dịch' columns
combined_df['Thuật Toán Và Ứng Dụng'] = combined_df['Thuật Toán Và Ứng Dụng'].fillna('').astype(str) + ', ' + combined_df['Chương Trình Dịch'].fillna('').astype(str)
combined_df['Thuật Toán Và Ứng Dụng'] = combined_df['Thuật Toán Và Ứng Dụng'].str.strip(', ')
combined_df.drop(columns=['Chương Trình Dịch'], inplace=True)
# Merge 'Lập Trình Thiết Bị Di Động' and 'Thực Hành Mạng' columns
combined_df['Lập Trình Thiết Bị Di Động'] = combined_df['Lập Trình Thiết Bị Di Động'].fillna('').astype(str) + ', ' + combined_df['Thực Hành Mạng'].fillna('').astype(str)
combined_df['Lập Trình Thiết Bị Di Động'] = combined_df['Lập Trình Thiết Bị Di Động'].str.strip(', ')
combined_df.drop(columns=['Thực Hành Mạng'], inplace=True)
# Merge 'Phân Tích Thiết Kế Yêu Cầu' and 'Phân Tích Thiết Kế Hệ Thống' columns
combined_df['Phân Tích Thiết Kế Yêu Cầu'] = combined_df['Phân Tích Thiết Kế Yêu Cầu'].fillna('').astype(str) + ', ' + combined_df['Phân Tích Thiết Kế Hệ Thống'].fillna('').astype(str)
combined_df['Phân Tích Thiết Kế Yêu Cầu'] = combined_df['Phân Tích Thiết Kế Yêu Cầu'].str.strip(', ')
combined_df.drop(columns=['Phân Tích Thiết Kế Hệ Thống'], inplace=True)
# Merge 'Phân Tích Thiết Kế Hướng Đối Tượng' and 'Xử Lý Ảnh' columns
combined_df['Phân Tích Thiết Kế Hướng Đối Tượng'] = combined_df['Phân Tích Thiết Kế Hướng Đối Tượng'].fillna('').astype(str) + ', ' + combined_df['Xử Lý Ảnh'].fillna('').astype(str)
combined_df['Phân Tích Thiết Kế Hướng Đối Tượng'] = combined_df['Phân Tích Thiết Kế Hướng Đối Tượng'].str.strip(', ')
combined_df.drop(columns=['Xử Lý Ảnh'], inplace=True)
# Merge 'Project 1' and 'Thực Tập Chuyên Môn' columns
combined_df['Project 1'] = combined_df['Project 1'].fillna('').astype(str) + ', ' + combined_df['Thực Tập Chuyên Môn'].fillna('').astype(str)
combined_df['Project 1'] = combined_df['Project 1'].str.strip(', ')
combined_df.drop(columns=['Thực Tập Chuyên Môn'], inplace=True)
# Merge 'Thiết Kế Cơ Sở Dữ Liệu', 'Hệ Quản Trị CSDL SQLServer', and 'Hệ Quản Trị CSDL Oracle' columns
combined_df.insert(
    combined_df.columns.get_loc('Hệ Quản Trị Csdl Sqlserver'),
    'Hệ Quản Trị CSDL SQLServer/Oracle',
    combined_df['Thiết Kế Cơ Sở Dữ Liệu'].fillna('').astype(str) + ', ' +
    combined_df['Hệ Quản Trị Csdl Sqlserver'].fillna('').astype(str) + ', ' +
    combined_df['Hệ Quản Trị Csdl Oracle'].fillna('').astype(str)
)
combined_df['Hệ Quản Trị CSDL SQLServer/Oracle'] = combined_df['Hệ Quản Trị CSDL SQLServer/Oracle'].str.strip(', ')
combined_df.drop(columns=['Thiết Kế Cơ Sở Dữ Liệu', 'Hệ Quản Trị Csdl Sqlserver', 'Hệ Quản Trị Csdl Oracle'], inplace=True)
# Merge 'Lập Trình Sử Dụng Api' and 'Công Nghệ Phần Mềm' columns
combined_df.insert(
    combined_df.columns.get_loc('Lập Trình Sử Dụng Api'),
    'Lập Trình Sử Dụng Api/Công Nghệ Phần Mềm',
    combined_df['Lập Trình Sử Dụng Api'].fillna('').astype(str) + ', ' +
    combined_df['Công Nghệ Phần Mềm'].fillna('').astype(str)
)
combined_df['Lập Trình Sử Dụng Api/Công Nghệ Phần Mềm'] = combined_df['Lập Trình Sử Dụng Api/Công Nghệ Phần Mềm'].str.strip(', ')
combined_df.drop(columns=['Lập Trình Sử Dụng Api', 'Công Nghệ Phần Mềm'], inplace=True)
# Xóa các hàng có nhiều hơn 40 cột trống
combined_df = combined_df.dropna(thresh=len(combined_df.columns) - 40)
# Xóa toàn bộ các cột không có giá trị nào
combined_df.dropna(axis=1, how='all', inplace=True)
# Xóa các cột có nhiều hơn 160 dòng trống
combined_df = combined_df.dropna(axis=1, thresh=len(combined_df) - 160)
# Xóa các cột không cần thiết
columns_to_drop = ['Khai Phá Dữ Liệu', 'Chuyên Đề Về Công Nghệ Phần Mềm', 'An Ninh Mạng']
combined_df.drop(columns=columns_to_drop, inplace=True)
# Fill missing values in specific columns with 0
columns_to_fill = ['Tiếng Anh Chuyên Ngành', 'Tiếng Anh B1', 'Thực Tập Tốt Nghiệp', 'Đồ Án Tốt Nghiệp']
combined_df[columns_to_fill] = combined_df[columns_to_fill].fillna(0)
# Lưu kết quả vào file mới
combined_df.to_excel('Combined_Results_2.xlsx', index=False)

print("Đã hoàn thành việc gộp hai file và quy đổi điểm.")
```
